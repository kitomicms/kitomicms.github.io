```python
import pandas as pd
import urllib
import requests
from bs4 import BeautifulSoup as bs
import datetime
```

```python
df1 = pd.DataFrame()
for k in range(1,11):

    url_hkjc = 'http://racing.hkjc.com/racing/Info/Meeting/Results/english/Local/20161112/ST/' + str(k)
    thepage_hkjc = urllib.urlopen(url_hkjc)
    soup_hkjc = bs(thepage_hkjc,"html.parser")
    records = []
    for tr in soup_hkjc.findAll("tr"):
        trs = tr.findAll("td")
        record = []
        try:
            for x in range(0,9):
                record.append(trs[x].text)
            records.append(record)
            for x in range()
        except:
            pass
    
    #resultsAll.append(records)
    
    df1 = pd.concat([df1,pd.DataFrame(data=records)])
```
