```python
import pandas as pd
import urllib
import requests
from bs4 import BeautifulSoup as bs
import datetime
import numpy as np
```

```python
df_full = pd.DataFrame()


for y in range(1,11):
    url = 'http://racing.hkjc.com/racing/Info/Meeting/Results/english/Local/20161112/ST/' + str(y)
    df = pd.read_html(url,header=0)
    df1 = pd.DataFrame(data=df[3])
    dfFirstRow = pd.DataFrame.transpose(pd.DataFrame(df1.columns))

    dfFirstRow.columns = df1.columns
    df1 = pd.concat([df1,dfFirstRow])
    
    fullList = list(['Place','Horse No.','Horse Name','Jockey','Trainer','Actual Wt.','Declar Horse Wt.','Draw','LBW'])

    for x in range(9,len(df1.columns)-2):
        fullList.append(x)

    fullList.append('Finish Time')
    fullList.append('Win Odds')
    df1.columns = fullList
    #['Place','Horse No.','Horse Name','Jockey','Trainer','Actual Wt.','Declar Horse Wt.','Draw','LBW',list(range(0,6)),'Finish Time','Win Odds']

    #df1.columns = fullList
    df1['Race'] = y
    df_full = pd.concat([df_full,df1[df1['Horse Name'].apply(lambda x: '(' in x )]])
```
