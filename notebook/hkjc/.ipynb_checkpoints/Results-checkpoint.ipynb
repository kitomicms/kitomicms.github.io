```python
import pandas as pd
import urllib
import requests
from bs4 import BeautifulSoup as bs
import datetime
```

```python
resultsAll = []
url_hkjc = 'http://racing.hkjc.com/racing/Info/Meeting/Results/english/Local/20161112/ST/' + str(1)
thepage_hkjc = urllib.urlopen(url_hkjc)
soup_hkjc = bs(thepage_hkjc,"html.parser")
records = []
for tr in soup_hkjc.findAll("tr"):
    trs = tr.findAll("td")
    record = []
    try:
        for x in range(0,15):
            record.append(trs[x].text)
        records.append(record)
    except:
        pass

#resultsAll.append(records)
df1 = pd.DataFrame(data=records)
```

```python
#df1 = pd.DataFrame(columns=range(0,17))



for y in range(2,11):

    url_hkjc = 'http://racing.hkjc.com/racing/Info/Meeting/Results/english/Local/20161112/ST/' + str(y)
    thepage_hkjc = urllib.urlopen(url_hkjc)
    soup_hkjc = bs(thepage_hkjc,"html.parser")
    records = []
    for tr in soup_hkjc.findAll("tr"):
        trs = tr.findAll("td")
        record = []
        try:
            for x in range(0,17):
                record.append(trs[x].text)
            records.append(record)
        except:
            pass
    
    #resultsAll.append(records)
    
    df1 = pd.concat([df1,pd.DataFrame(data=records)])
    
```
